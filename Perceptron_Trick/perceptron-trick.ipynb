```python
from sklearn.datasets import make_classification # type: ignore
import numpy as np # type: ignore
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt # type: ignore
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1]) #initialize the weights with 1
    
    lr = 0.1
    epochs = 1000
    
    for i in range(epochs): 
        j = np.random.randint(0,100) #choosing a random student everytime
        y_hat = step(np.dot(X[j],weights)) #predicted value
        weights = weights + lr*(y[j]-y_hat)*X[j] #update the weights
        
    return weights[0],weights[1:] #weights[0] : intercept and weights[1:] : all the coeffecients
        
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y) #call the perceptron function and store the values in intercept and coeffecient
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
